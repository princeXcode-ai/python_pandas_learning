```python
df['Total marks'] = df['Science'] + df['History'] + df['English'] + df["Maths"]
```

```python
df
```

```python
df['Percentage'] = df["Total marks"]/ 4
```

```python
df
```

```python
def grade(per):
    if per >= 90:
        return "A+"
    elif per >= 80:
        return "A"
    elif per >= 70:
        return "B"
    elif per >= 60:
        return "C"
    elif per >= 50:
        return "D"
    elif per >= 40:
        return "E"
    else:
        return "F"

df["Grade"] = df["Percentage"].apply(grade)
         
```
